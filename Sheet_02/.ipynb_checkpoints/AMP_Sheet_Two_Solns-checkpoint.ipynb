## Physics Question One

Consider the Schrodinger Wave Equation for a partical moving in a one-dimensional universe against a potential $V(x)$

$$
i \hbar \frac{\partial \Psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + V(x)\Psi
$$

Prove that the following equation is a solution: 

$$
\Psi(x,t) = A e^{i(kx-wt)}
$$

**Hint**: _Think about how "proving this is a solution" is done otherwise you might get to the end and not realise it._


To prove we have a solution, if we get a physical law out, this implies that the solution is correct. 

### First step: 
Get the derivatives.

We need: 
$$
\frac{\partial^{2} \Psi}{\partial x^{2}} \ \ \text{and} \ \ \frac{\partial \Psi}{\partial t}
$$

$$
\frac{\partial^{2} \Psi}{\partial x^{2}} = \frac{\partial^{2}}{\partial x^{2}} \bigg( A e^{i(kx-\omega t)} \bigg)
$$

$$
\frac{\partial^{2} \Psi}{\partial x^{2}} = Ae^{-\omega t}\frac{\partial^{2}}{\partial x^{2}} \big( e^{i(kx)} \big)
$$

$$
\frac{\partial^{2} \Psi}{\partial x^{2}} = Ae^{-\omega t}ik \frac{\partial}{\partial x} \big( e^{i(kx)} \big)
$$

$$
\frac{\partial^{2} \Psi}{\partial x^{2}} = Ae^{-\omega t} (ik)^{2} e^{i(kx)}
$$

$$
\frac{\partial^{2} \Psi}{\partial x^{2}} = (ik)^{2} A e^{i(kx- \omega t)}
$$

Woohoo! Now for the time derivative:

$$
\frac{\partial \Psi}{\partial t} = \frac{\partial }{\partial t} \big( A e^{i(kx- \omega t)} \big)
$$

$$
\frac{\partial \Psi}{\partial t} = A e^{ikx} \frac{\partial }{\partial t} \big( e^{-i \omega t} \big)
$$

$$
\frac{\partial \Psi}{\partial t} = A e^{ikx} (-i \omega)  e^{-i \omega t} 
$$

$$
\frac{\partial \Psi}{\partial t} = (-i \omega )A e^{i(kx- \omega t)}
$$

### Second step: 
Substitute our derivatives and the wave equation into the Schrodinger Wave Equation

$$
i \hbar \frac{\partial \Psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + V(x)\Psi
$$

$$
i \hbar (-i\omega)A e^{i(kx-wt)} = -\frac{\hbar^2}{2m}(ik)^{2} A e^{i(kx-\omega t)}  + V(x)A e^{i(kx-\omega t)}
$$

Factor out $A e^{i(kx-\omega t)}$

$$
i \hbar (-i\omega) = -\frac{\hbar^2}{2m}(ik)^{2}  + V(x)
$$

Simplify our imaginary numbers:

$$
\hbar \omega = \frac{\hbar^2}{2m}(k)^{2}  + V(x)
$$

And we know that:

$$
\omega = 2 \pi f \ \ \ \text{and} \ \ \ p = hk \ \ \ \text{and} \ \ \ \hbar = \frac{h}{2\pi}
$$

$$
E = h f = \frac{(\hbar k)^{2}}{2m}  + V(x)
$$

Here we have a physical law of the universe, E=hf, which could only become self-evident because we have solved the equation correctly (if you want to know more about why this approach is true, ask your lecturers).

## Mathematics: Question One

Consider the force field:

$$
\mathbf{F}(x, y) = x^2\mathbf{i} âˆ’ xy\mathbf{j}
$$

A particle is moved by this field along a straight line path $C$ defined as:

$$
C:  \begin{bmatrix} 1 \\ 0 \end{bmatrix} \rightarrow  \begin{bmatrix} 4 \\ 1 \end{bmatrix} 
$$

Find the amount of work done on the particle. 

The first thing to consider is what kind of field this is. 

Of the two kinds of fields we do line-integrals on, this is the **vector** kind. I'll endeavour to always indicate which variables are vectors with a bold font.

We also want to figure out what the _work_ is. From Physical Modelling we can recall that:

$$
W = F s
$$

Which would be perfectly applicable if we were working with a constant force in a conservative field (more on conservative fields at a later date). But, we don't yet know if this field is conservative and the force evidently changes with position!

So, let's make changes accordingly:

$$
W = \int \mathbf{F} \cdot d\mathbf{s}
$$

**Why does this make sense?** Because our force is given by a vector field. A force field is a vector of force that is dependent on what $x$ and $y$ coordinate we input into the function. Also, remember when we were asked repetitively to define the difference between scalars and vectors? We should all know off by heart that a displacement and force are **vector** quantities by now (if you forgot, all good, I've just reminded you!).

Another one of our trusty line integral formulas will be used to pull us out of this problem:

$$
\int \mathbf{F} \cdot d\mathbf{s} = \int \mathbf{F}(\mathbf{r}) \cdot \mathbf{r^{'}}(t) \ dt
$$

### Let's paramaterise the path: 

We're travelling in a straight line, so let's do our linear algebra technique. 

$$
(1-t)  \begin{bmatrix} 1 \\ 0 \end{bmatrix} +  \ t  \begin{bmatrix} 4 \\ 1 \end{bmatrix}
$$

$$
= \begin{bmatrix} 1-t \\ 0 \end{bmatrix} +  \begin{bmatrix} 4t \\ t \end{bmatrix}
$$

$$
\mathbf{r}(t) = \begin{bmatrix} x(t) \\ y(t) \end{bmatrix} = \begin{bmatrix} 1+3t \\ t\end{bmatrix}
$$

We know that our vector $\mathbf{i}$ and $\mathbf{j}$ are unit vectors for the $x$ and $y$ axes, respectively. So we can hence express the vectorised path as:

$$
\mathbf{r}(t) 
= \begin{bmatrix} x(t) \\ y(t) \end{bmatrix} 
\begin{bmatrix} \hat{\imath} & \hat{\jmath} \end{bmatrix} 
= \begin{bmatrix} 1+3t \\ t \end{bmatrix}
\begin{bmatrix} \hat{\imath} & \hat{\jmath} \end{bmatrix}
$$

Which means that our parameterised functions of $x$ and $y$ are:

$$
x(t) = 1 + 4t \qquad y(t) = t
$$

### Inserting our path into our field

So the next part of the problem is to insert the path into the field, as per:

$$
\int \mathbf{F}(\mathbf{r}) \cdot \mathbf{r^{'}}(t) \ dt
$$

So what we're doing is this part:

$$
\mathbf{F}(\mathbf{r})
$$

Just to be clear, let's also express the force field as a vector too:

$$
\mathbf{F}(x,y) = \begin{bmatrix} x^2 \\ -xy \end{bmatrix}
$$

Inserting our $x(t)$ and $y(t)$ into our field:

$$
\mathbf{F}(\mathbf{r}(t)) 
= \begin{bmatrix} \big( x(t)\big)^2 \\ -\big(x(t) \ y(t)\big) \end{bmatrix}
= \begin{bmatrix} \bigg(1+3t\bigg)^2 \\ -\bigg((1+3t) \ t\bigg) \end{bmatrix}
= \begin{bmatrix} 9t^2 + 6t + 1 \\   -3t^2 - t \end{bmatrix}
$$

### What's $ \mathbf{r^{'}}(t)$ ?

It's simply the derivative of the parameterised path!

$$
\mathbf{r^{'}}(t) 
= \frac{d}{dt} \
\mathbf{r}(t) 
= \frac{d}{dt} \begin{bmatrix} 1+3t \\ t \end{bmatrix}
= \begin{bmatrix} 3 \\ 1 \end{bmatrix}
$$

I'm also going to transpose our matrix so that our dot product (below) will be valid. 

$$
 \begin{bmatrix} 3 \\ 1 \end{bmatrix} ^T =  \begin{bmatrix} 3 & 1 \end{bmatrix}
$$

You might wonder whether I'm allowed to just do that, and I sort of don't know? But, it gets the right answer. I also arbitrarily chose the orientation of my path derivative's ($ \mathbf{r^{'}}(t)$ ) matrix.

### Dot product with matrices under the integral sign!

$$
\int \mathbf{F}(\mathbf{r}) \cdot \mathbf{r^{'}}(t) \ dt \\
= \int^{1}_{0} \begin{bmatrix} 9t^2 + 6t + 1 \\   -3t^2 - t \end{bmatrix} 
\cdot \begin{bmatrix} 3 & 1 \end{bmatrix} \cdot dt \\ 
$$

$$ 
= \int^{1}_{0} 3(9t^2+6t+1) -t - 3t^2  \ \cdot dt
$$

$$
= \int^{1}_{0} 24t^2+17t+3 \ \cdot dt
$$

$$
= \bigg[ 8t^3 + \frac{17}{2}t^2 + 3t  \bigg]^{1}_{0}
$$

$$
= 8 + \frac{17}{2} + 3 
$$

$$
= 19.5 \ \text{(Joules)}
$$
