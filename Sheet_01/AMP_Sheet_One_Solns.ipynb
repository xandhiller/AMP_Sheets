## Parameterising our path

$$ x =r\cos(x) $$

$$ y = r \sin(x) $$

$ds$ is a _path_.



**Arc Length Formula:**

$$ ds = \sqrt{dy^{2}+dx^{2}} $$

$$ ds = \sqrt{ \bigg( {\frac{dy}{dx}}\bigg)^{2}+1}  \ \ dx $$

_Doesn't really work with what we're doing (a bit complicated)._



**Examining the arc length of a circle:**

$$ s = \theta r $$

$$ \therefore \quad ds =  dr \ d\theta $$


**Inserting into this:**

$$ \int_{C}(3x+2xy^2) \ ds $$

** We get:**

$$ \int \int \big( 3r\cos \theta + 2r \cos \theta  \ r^2 \sin^{2} \theta \ \big) r \ dr \ d\theta $$

** Inserting the bounds: **
$$ \int^{\frac{\pi}{2}}_{0} \int^{\sqrt{3}}_{0}  3r^2 \cos \theta + 2r^4 \cos \theta \sin^{2} \theta \ \ dr \ d\theta $$

## Cranking out the integral
$$ 
\int^{\frac{\pi}{2}}_{0} \int^{\sqrt{3}}_{0}  3r^2 \cos \theta + 2r^4 \cos \theta \sin^{2} \theta \ \ dr \ d\theta 
$$

$$ \int^{\frac{\pi}{2}}_{0} \bigg[  r^3\cos \theta + \frac{2r^5}{5} \cos \theta  \sin^{2} \theta  \ d\theta  \bigg]^\sqrt{3}_{0} $$


$$ 
\int^{\frac{\pi}{2}}_{0} 3\sqrt{3}\cos \theta + \frac{18\sqrt{3}}{5} \cos \theta  \sin^{2} \theta \ d\theta  
$$

Separating the integral:

$$ 
 \bigg[ 3\sqrt{3}\sin \theta \bigg]^{\frac{\pi}{2}}_{0}  \ +  \ \frac{18\sqrt{3}}{5} \int^{\frac{\pi}{2}}_{0} \cos \theta  \sin^{2} \theta \ d\theta  
$$

$$ 
3\sqrt{3}  \ +  \ \frac{18\sqrt{3}}{5} \int^{\frac{\pi}{2}}_{0} \cos \theta  \sin^{2} \theta \ d\theta  
$$

_Now for some $u$ substitution:_

$$ u = \sin\theta $$

$$ \frac{du}{d\theta} = \cos \theta $$

$$ \frac{du}{\cos \theta} =  d\theta $$

_Putting it into the integral:_

$$ 3\sqrt{3} \ \ + \ \ \frac{18\sqrt{3}}{5} \int^{\frac{\pi}{2}}_{0} \cos \theta  \sin^{2} \theta  \ d\theta  $$

$$ 3\sqrt{3}  \ \ + \ \ \frac{18\sqrt{3}}{5} \int^{\frac{\pi}{2}}_{0} \cos \theta \ \  u^{2} \  \frac{du}{\cos \theta} $$

$$ 3\sqrt{3} \ \ + \ \ \frac{18\sqrt{3}}{5} \int^{\frac{\pi}{2}}_{0} u^{2} \  du $$

$$ 3\sqrt{3}  \ \ + \ \ \frac{18\sqrt{3}}{5}  \bigg[ \frac{u^{3}}{3} \bigg]^{\frac{\pi}{2}}_{0} $$

_Chuck $u$ back in:_

$$ 3\sqrt{3}  \ \ + \ \ \frac{18\sqrt{3}}{5}  \bigg[ \frac{{\sin^{3}\theta}}{3} \bigg]^{\frac{\pi}{2}}_{0} $$

$$ 3\sqrt{3}  \ \ + \ \ \frac{18\sqrt{3}}{5} \frac{1}{3}  $$

$$ 3\sqrt{3}  \ \ + \ \ \frac{6\sqrt{3}}{5}   $$

$$ = \frac{21\sqrt{3}}{5} $$

# Mathematics: Question Two

Evaluate the integral:
$$ \int \big(x^2-y \big) \ dx \ + \  \big( y^2+x  \big) \ dy \ $$

Along $C$, the straight line from (0,1) $\rightarrow$ (1,2)

## Parameterise!
This might be a new trick, but I bloody love this one. Works only for straight lines (as it uses _linear_ algebra).

$$ \big( 1-t \big) \begin{bmatrix} 0 \\ 1 \end{bmatrix}\ + \ t \ \begin{bmatrix} 1 \\ 2 \end{bmatrix} $$

Hence:

$$ x = 0 + t $$ 

$$ y \ = \ 2t + 1 - t \ = \ t +1 $$

But we need to calculate the squares of $x$ and $y$ and differentials ($dx$ and $dy$) in order to crank out the integral:

$$ 
x^2 = t^2 
$$ 

$$
y^2 = (t+1)^2  = t^2 + 2t + 1
$$

$$
\frac{dx}{dt} = 1 \qquad {dx} =  dt
$$

$$
\frac{dy}{dt} = 1 \qquad {dy} =  {dt}
$$




## Crank out the integral

$$ \int \big(x^2-y \big) \ dx \ + \  \big( y^2+x  \big) \ dy \ $$

Substitute in our values:

$$ 
\int_0^1 \big( t^2 - (t +1) \big) \ dt \ + \  \big( t^2 + 2t + 1 + t  \big) \ {dt} \ 
$$

Expanding:

$$ 
\int_0^1 \big( t^2 - t -1  +   t^2 + 3t + 1 \big) \ {dt} \ 
$$

Simplifying:

$$ 
\int_0^1 \big( 2t^2 + 2t \big) \ dt 
$$

Evaluating the integral:
$$
\big[ \frac{2t^3}{3} + \frac{2t^2}{2} \big]^{1}_{0} = \frac{2}{3} + 1 = \frac{5}{3}
$$

# Physics: Question One

(a) What is the name of the following equation?

$$ i\hbar \frac{d\phi}{dt} \ = \ E \ \phi(t) $$

(b) Solve the differential equation for $\phi(t)$:

$$ i\hbar \frac{d\phi}{dt} \ = \ E \ \phi(t) $$

Such that:

$  i=\sqrt{-1}$

$ h  = 6.626 \times 10^{-34} \ m^{2}kg/s $

$ \hbar = h \ / \ 2\pi  $

$E \ \ $ is a constant

---
## Name of the equation:
It is a variant of the _Schrodinger Equation_. 

---

## A first order differential equation! Old tricks should work:

$$ i\hbar \frac{d\phi}{dt} \ = \ E \ \phi(t) $$

$$  \frac{d\phi}{\phi(t)} \ = \ \frac{E \ dt}{i\hbar} $$

Clean up $i$:

$$  \frac{d\phi}{\phi(t)} \ = \ \frac{E \ dt}{i\hbar} \frac{i}{i} $$

$$  \frac{d\phi}{\phi(t)} \ = \ \frac{-i \ E \ dt}{\hbar} $$

Integrate both sides:

$$  \int \frac{d\phi}{\phi(t)} \ = \ \int \frac{-i \ E}{\hbar}dt $$

$$  \ln[  \phi(t)  ] + C_1 \ = \ \frac{-i \ E}{\hbar}t + C_2 $$

Make both sides the exponent of $e$:

$$  e^{\ln[\phi(t)] + C_1} \ = \ e^{\frac{-i \ E}{\hbar}t + C_2} $$

Use exponent rules to simplify:

$$ A = \frac{e^{C_2}}{e^{C_1}} $$

$$ \phi(t) \ = \ A \ e^{\frac{-i \ E}{\hbar}t} $$
